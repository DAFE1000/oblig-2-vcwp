```python
"""Skript for å beregne tilnærmede ekstremalpunkter ved hjelp av 
Newtons metode på den deriverte f'(x)
"""
# Importerer biblioteker
import numpy as np

# Definerer den deriverte og den andrederiverte
def df(x):
    return (np.arctan(x) - (4/(x**2 + 1)))
    
def ddf(x):
    return (1/(x**2 + 1)) + (8*x/(x**2 + 1)**2)

# Bestemmer x_0
x = 1

# Presisjon
pres = 1e-5

# Inititerer tallet på iterasjoner
iterasjoner = 0

# Lager en variabel for "gammel" x-verdi
x_gml = 100

while abs(x-x_gml) > pres:
    # Kopierer den gamle x
    x_gml = x
    # Iterasjonsskjema
    x = x - df(x) / ddf(x)
    # Teller iterasjoner
    iterasjoner = iterasjoner + 1

# Skriver svaret til skjerm
print(f'x = {x:.5f}')

# Skriver tallet på iterasjoner til skjerm
print(f'interasjoner = {iterasjoner}')
```
