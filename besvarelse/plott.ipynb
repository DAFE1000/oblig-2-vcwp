```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(-x/4) * np.arctan(x)

# Verdier for plotting av graf
x = np.linspace(-1, 15, 1000)
y = f(x)

# Toppunkt
x_topp = 1.69071
y_topp = f(x_topp)

plt.plot(x, y, label='e^(-x/4)*tan^-1(x)')
plt.plot(x_topp, y_topp, 'ro', markersize=6, label='toppunkt')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('f(x) = e^(-x/4)*tan^-1(x)')
plt.legend()
plt.grid(True)
plt.show()
```
